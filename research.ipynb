```python
import requests
import pandas as pd

def fetch_papers(query, from_date, to_date, page=1, per_page=10):
    url = (
        f"https://api.openalex.org/works?"
        f"search={query}&"
        f"sort=relevance_score:desc&"
        f"filter=from_publication_date:{from_date},to_publication_date:{to_date}&"
        f"per-page={per_page}&page={page}"
    )

    response = requests.get(url)

    if response.status_code == 200:
        return response.json()["results"]
    else:
        raise Exception(f"API Error: {response.status_code}, {response.text}")

def get_abstract(abstract_index):
    """
    Decompress the abstract from its inverted index format.
    """
    if not abstract_index:
        return "Abstract not available"

    # Create a list to store the words in the correct order
    abstract_length = max([max(positions) for positions in abstract_index.values()])
    abstract_words = [""] * (abstract_length + 1)

    # Fill the abstract_words list based on the positions
    for word, positions in abstract_index.items():
        for position in positions:
            abstract_words[position] = word

    # Join the words to form the full abstract
    abstract = " ".join(abstract_words).strip()
    return abstract

# Example Usage
query = "Hair Loss treatment"
from_date = "2022-01-01"
to_date = "2025-01-01"
papers = fetch_papers(query, from_date, to_date)

# Prepare data for saving to Excel
paper_data = []

for paper in papers:
    title = paper["display_name"]
    authors = ", ".join([auth["author"]["display_name"] for auth in paper["authorships"]])
    publication_date = paper.get("publication_date", "N/A")
    doi = paper.get("doi", "Not available")
    abstract = get_abstract(paper.get("abstract_inverted_index"))

    # Collect the data in a dictionary
    paper_data.append({
        "Title": title,
        "Authors": authors,
        "Publication Date": publication_date,
        "DOI": doi,
        "Abstract": abstract
    })

# Convert the data to a DataFrame
df = pd.DataFrame(paper_data)

# Save the DataFrame to an Excel file
# excel_filename = "research_papers.xlsx"
# df.to_excel(excel_filename, index=False)

# print(f"Data saved to {excel_filename}")
df.head()
```

```python
import requests
import os
import pandas as pd

# Function to download the PDF using DOI
def download_pdf(doi, folder_path):
    """
    Download PDF using DOI link and save it to the folder path.
    """
    # Construct the DOI URL to download the PDF
    url = f"https://doi.org/{doi}"

    try:
        # Make the request to download the PDF
        response = requests.get(url, headers={"Accept": "application/pdf"})
        
        # If response is successful and contains a PDF
        if response.status_code == 200 and 'application/pdf' in response.headers.get('Content-Type', ''):
            # Generate a valid filename using DOI (replace / with _)
            file_name = f"{doi.replace('/', '_')}.pdf"
            file_path = os.path.join(folder_path, file_name)

            # Save the PDF file to the specified folder
            with open(file_path, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {file_name}")
        else:
            print(f"PDF not found for DOI: {doi}")
    
    except Exception as e:
        print(f"Error downloading {doi}: {e}")

# Folder to save PDFs
folder_path = "downloaded_papers"
os.makedirs(folder_path, exist_ok=True)

# Example: Assuming df already contains the DOI column
# Loop through the rows of the DataFrame to download PDFs based on DOI
for index, row in df.iterrows():
    doi = row['DOI']
    if pd.notna(doi):  # Check if DOI is not empty
        download_pdf(doi, folder_path)

```
